```python
import pandas as pd
```

```python
import numpy as np
```

```python
df = pd.read_csv('https://github.com/YBIFoundation/Dataset/raw/main/Movies%20Recommendation.csv')
```

```python
df_features= df[['Movie_Genre','Movie_Keywords','Movie_Tagline', 'Movie_Cast','Movie_Director']].fillna('')
```

```python
df_features.shape
```

```python
df_features
```

```python
x=df_features['Movie_Genre']+' '+df_features['Movie_Keywords']+' '+df_features['Movie_Tagline']+df_features['Movie_Cast']+df_features['Movie_Director']
```

```python
#x.shapeGet feature Text conversion to Tokens
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf = TfidfVectorizer()
```

```python
x=tfidf.fit_transform(x)
```

```python
x.shape
```

```python
print(x)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score = cosine_similarity(x)
```

```python
similarity_score
```

```python
similarity_score.shape
```

```python
Favorite_Movie_Name =input('Enter your favorite movie name: ')
```

```python
All_Movies_Title_List = df['Movie_Title'].tolist()
```

```python
import difflib
```

```python
Movie_Recomendation= difflib.get_close_matches(Favorite_Movie_Name, All_Movies_Title_List)
```

```python
Close_Match = Movie_Recomendation[0]
print(Close_Match)
```

```python
Index_of_Close_Match_Movie =df[df.Movie_Title == Close_Match]['Movie_ID'].values[0]
```

```python
#Getting a list of similar movies
Recommendation_Score=list(enumerate(similarity_score[Index_of_Close_Match_Movie]))
print(Recommendation_Score)
```

```python
len(Recommendation_Score)
```

Get all movies sort based on recomendation score wrt Favorite movie

```python
#sorting the movies based on their similarity score
Sorted_Similar_Movies = sorted(Recommendation_Score, key =lambda x:x[1], reverse=True)
Sorted_Similar_Movies
```

```python
#print the name of the similar movies based on the index
print('Top 30 Movies Suggested for You : \n')

i=1

for movie in Sorted_Similar_Movies:
  index =movie[0]
  title_from_index =df[df.index==index]['Movie_Title'].values[0]
  if (i<31):
    print(i, '.', title_from_index)
    i+=1
```
